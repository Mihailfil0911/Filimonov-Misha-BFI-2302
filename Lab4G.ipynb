```python

import time
import random
import bisect
from collections import deque

# Генерация случайных данных
def generate_data(size, start=0, end=1000):
    return sorted(random.sample(range(start, end), size))

# Бинарный поиск (итеративный) с каждым шагом сужает диапозон поиска на половину
def binary_search(arr, target):
    left, right = 0, len(arr) - 1  # Инициализируем левую и правую границы поиска
    while left <= right:  # Пока левая граница не перекрыла правую
        mid = (left + right) // 2  # Вычисляем средний индекс
        if arr[mid] == target:  # Если элемент в середине равен target
            return mid  # Возвращаем индекс
        elif arr[mid] < target:  # Если элемент в середине меньше target
            left = mid + 1  # Сдвигаем левую границу вправо
        else:  # Если элемент в середине больше target
            right = mid - 1  # Сдвигаем правую границу влево
    return -1  # Если target не найден, возвращаем -1

# Узел бинарного дерева (связь с неболее двумя потомками)
class TreeNode:
    def __init__(self, value):
        self.value = value  # Значение узла
        self.left = None  # Левый потомок
        self.right = None  # Правый потомок

# Бинарное дерево (с итеративной вставкой) поиск по заданному ключу, если имеется ключ, то заверш, иначе вставка (правая чаcть >; левая часть <)
class BinaryTree:

    def __init__(self):

        self.root = None  # Корень дерева

    def insert(self, value):
        new_node = TreeNode(value)  # Создаем новый узел
        if not self.root:  # Если дерево пустое
            self.root = new_node  # Новый узел становится корнем
            return

        current = self.root  # Начинаем с корня
        while True:  # Цикл до тех пор, пока не найдем место для вставки
            if value < current.value:  # Если значение меньше текущего узла
                if current.left is None:  # Если левого потомка нет
                    current.left = new_node  # Вставляем новый узел слева
                    break  # Завершаем вставку
                current = current.left  # Переходим к левому потомку
            else:  # Если значение больше или равно текущему узлу
                if current.right is None:  # Если правого потомка нет
                    current.right = new_node  # Вставляем новый узел справа
                    break  # Завершаем вставку
                current = current.right  # Переходим к правому потомку

    def search(self, value):
        current = self.root  # Начинаем с корня
        while current:  # Пока не дошли до конца дерева
            if current.value == value:  # Если значение текущего узла равно искомому
                return True  # Возвращаем True
            elif value < current.value:  # Если искомое значение меньше текущего
                current = current.left  # Переходим к левому потомку
            else:  # Если искомое значение больше текущего
                current = current.right  # Переходим к правому потомку
        return False  # Если значение не найдено, возвращаем False

# Фибоначчиев поиск (быстрое нахождение нужного значения в отсорт списке, выбирает точки сравнения в списке, основ на числ Фибоначи [принцип суммы пред двух])
def fibonacci_search(arr, target):
    size = len(arr)
    # Шаг 1: находим наибольшее число Фибоначчи, меньшее или равное размеру массива
    fib2, fib1 = 0, 1  # Начальные значения чисел Фибоначчи
    fibM = fib1 + fib2  # Следующее число Фибоначчи
    while fibM < size:  # Пока число Фибоначчи меньше размера массива
        fib2, fib1 = fib1, fibM  # Сдвигаем числа Фибоначчи
        fibM = fib1 + fib2  # Вычисляем следующее число Фибоначчи

    offset = -1 # Смещение относительно начала массива
    while fibM > 1:  # Пока текущее число Фибоначчи больше 1
        i = min(offset + fib2, size - 1)  # Вычисляем индекс для проверки
        if arr[i] < target:  # Если элемент в arr[i] меньше target
            fibM, fib1, fib2 = fib1, fib2, fib1 - fib2  # Сдвигаем числа Фибоначчи
            offset = i  # Обновляем смещение
        elif arr[i] > target:  # Если элемент в arr[i] больше target
            fibM, fib1, fib2 = fib2, fib1 - fib2, fib2 - fib1  # Сдвигаем числа Фибоначчи
        else:  # Если элемент в arr[i] равен target
            return i  # Возвращаем индекс
    if fib1 and offset + 1 < size and arr[offset + 1] == target:  # Проверяем последний элемент
        return offset + 1  # Возвращаем индекс
    return -1  # Если target не найден, возвращаем -1

# Интерполяционный поиск (предсказание местонахождения элемента)
def interpolation_search(arr, target):

    low, high = 0, len(arr) - 1  # Инициализируем левую и правую границы поиска
    while low <= high and arr[low] <= target <= arr[high]:  # Пока границы не пересеклись и target находится в диапазоне
        # Вычисляем предполагаемую позицию target на основе интерполяции
        pos = low + ((target - arr[low]) * (high - low) // (arr[high] - arr[low]))
        if arr[pos] == target:  # Если элемент в pos равен target
            return pos  # Возвращаем индекс
        if arr[pos] < target:  # Если элемент в pos меньше target
            low = pos + 1  # Сдвигаем левую границу вправо
        else:  # Если элемент в pos больше target
            high = pos - 1  # Сдвигаем правую границу влево
    return -1  # Если target не найден, возвращаем -1

# Функция для измерения времени работы алгоритмов
def measure_time(search_func, arr, target):
    start_time = time.time()  # Записываем время начала
    result = search_func(arr, target)  # Выполняем функцию поиска
    end_time = time.time()  # Записываем время окончания
    return result, end_time - start_time  # Возвращаем результат и время

# Генерация данных и тестирование
size = 1000  # Размер данных
data = generate_data(size)  # Генерируем отсортированные данные
random.shuffle(data)  # Перемешивание данных перед вставкой в дерево, чтобы дерево не было вырожденным
target = random.choice(data)  # Выбираем случайный элемент из данных для поиска

# Создание и наполнение бинарного дерева
tree = BinaryTree()  # Создаем бинарное дерево
for num in data:  # Итерируемся по данным
    tree.insert(num)  # Вставляем каждое число в дерево

# Замер времени работы
print("Built-in bisect:", measure_time(lambda arr, x: bisect.bisect_left(arr, x), data, target))

print("Binary Search:", measure_time(binary_search, data, target))

print("Binary Tree Search:", measure_time(lambda arr, x: tree.search(x), data, target))

print("Fibonacci Search:", measure_time(fibonacci_search, data, target))


print("Interpolation Search:", measure_time(interpolation_search, data, target))


```

```python

import random
from typing import List, Optional

class HashTableBase:
    def __init__(self, size: int):
        self.size = size
        self.table = [None] * size  # Создаем таблицу, заполненную None
    def _hash(self, key: str) -> int:
        # Простой хеш-функция: сумма ASCII-значений символов ключа % размер таблицы
        return sum(ord(c) for c in key) % self.size
    def insert(self, key: str, value: any) -> None:
        raise NotImplementedError()
    def search(self, key: str) -> Optional[any]:
        raise NotImplementedError()
    def delete(self, key: str) -> None:
        raise NotImplementedError()


# 1. Простое рехэширование (линейное пробирование) изменение размера хеш-таблицы, распред её элемент для ум коллизий (хешы в разной послед) 
# и обеспеч работы структуры данных
class LinearProbingHashTable(HashTableBase):
    def insert(self, key: str, value: any) -> None:
        index = self._hash(key) # Вычисляем исходный индекс
        while self.table[index] is not None and self.table[index][0] != key:
            # Если ячейка занята и ключ не совпадает
            index = (index + 1) % self.size  # Линейное пробирование: переходим к следующей ячейке
        self.table[index] = (key, value)  # Вставляем пару ключ-значение

    def search(self, key: str) -> Optional[any]:

        index = self._hash(key) # Вычисляем исходный индекс
        while True:
            if self.table[index] is None:
                return None # Ключ не найден (дошли до пустой ячейки)
            if self.table[index][0] == key:
                return self.table[index][1]  # Найдено значение
            index = (index + 1) % self.size  # Продолжаем поиск

    def delete(self, key: str) -> None:

        index = self._hash(key) # Вычисляем исходный индекс
        while True:
            if self.table[index] is None:
                return  # Ключ не найден
            if self.table[index][0] == key:
                self.table[index] = None  # Удаляем элемент
                return
            index = (index + 1) % self.size  # Продолжаем поиск


# 2. Рехэширование с помощью псевдослучайных чисел (использ послед чисел для сниж сред времени на помещ 1-ого элем в таб и на поиск элем в ней)
class RandomProbingHashTable(HashTableBase):
    def insert(self, key: str, value: any) -> None:

        index = self._hash(key)  # Вычисляем исходный индекс
        step = random.randint(1, self.size - 1)  # Случайный шаг (от 1 до size-1)
        while self.table[index] is not None and self.table[index][0] != key:
            index = (index + step) % self.size  # Случайное пробирование
        self.table[index] = (key, value)  # Вставляем пару ключ-значение

    def search(self, key: str) -> Optional[any]:
        index = self._hash(key)  # Вычисляем исходный индекс
        step = random.randint(1, self.size - 1)  # Случайный шаг
        while True:
            if self.table[index] is None:
                return None  # Ключ не найден
            if self.table[index][0] == key:
                return self.table[index][1]  # Найдено значение
            index = (index + step) % self.size  # Продолжаем поиск

    def delete(self, key: str) -> None:
        index = self._hash(key)  # Вычисляем исходный индекс
        step = random.randint(1, self.size - 1)  # Случайный шаг
        while True:
            if self.table[index] is None:
                return  # Ключ не найден
            if self.table[index][0] == key:
                self.table[index] = None  # Удаляем элемент
                return
            index = (index + step) % self.size  # Продолжаем поиск


# 3. Метод цепочек (результ выз 1-ого метода становится контекстом для след)
class ChainingHashTable(HashTableBase):
    def __init__(self, size: int):

        super().__init__(size)
        self.table = [[] for _ in range(size)]  # Каждая ячейка - список (связанный [одна структура из связанных узлов])

    def insert(self, key: str, value: any) -> None:
        index = self._hash(key)  # Вычисляем индекс
        for i, (k, v) in enumerate(self.table[index]):
            if k == key:
                self.table[index][i] = (key, value)  # Обновляем значение, если ключ уже существует
                return
        self.table[index].append((key, value))  # Добавляем новый элемент в цепочку

    def search(self, key: str) -> Optional[any]:
        index = self._hash(key)  # Вычисляем индекс
        for k, v in self.table[index]:
            if k == key:
                return v  # Найдено значение
        return None  # Ключ не найден

    def delete(self, key: str) -> None:
        index = self._hash(key)  # Вычисляем индекс
        for i, (k, v) in enumerate(self.table[index]):
            if k == key:
                del self.table[index][i]  # Удаляем элемент из цепочки
                return


# Тестирование
if __name__ == "__main__":
    # Тестовая функция для всех
    pass
```

### Задание №3

```python
def print_board(board):
    for row in board:
        print(" ".join("F" if cell else "." for cell in row))  # Преобразуем True в 'F', False в '.', объединяем в строку
    print("\n")


# Функция для проверки безопасности размещения ферзя в позиции (row, col)
def is_safe(board, row, col, n):
    # Проверяем столбец
    for i in range(row):  # Проверяем только строки выше текущей
        if board[i][col]:  # Если в этом столбце выше есть ферзь
            return False  # Небезопасно

    # Проверяем левую верхнюю диагональ
    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):  # Идем по диагонали вверх и влево
        if board[i][j]:  # Если на диагонали есть ферзь
            return False  # Небезопасно

    # Проверяем правую верхнюю диагональ
    for i, j in zip(range(row, -1, -1), range(col, n)):  # Идем по диагонали вверх и вправо
        if board[i][j]:  # Если на диагонали есть ферзь
            return False  # Небезопасно

    return True  # Если все проверки пройдены, безопасно


# Рекурсивная функция для размещения n ферзей
def solve_n_queens(board, row, n): # принцип разделяй и властвуй
    if row == n:  # Если мы разместили ферзей во всех строках
        print_board(board)  # Печатаем доску с решением
        return True  # Решение найдено

    # Пробуем разместить ферзя в каждой колонке текущей строки
    for col in range(n):
        if is_safe(board, row, col, n):  # Если безопасно размещать ферзя в данной позиции
            board[row][col] = True  # Размещаем ферзя

            if solve_n_queens(board, row + 1, n):  # Рекурсивно вызываем функцию для следующей строки
                return True  # Если рекурсивный вызов вернул True, значит, решение найдено

            board[row][col] = False  # Если рекурсивный вызов вернул False, отменяем размещение ферзя (backtrack)

    return False  # Если не удалось разместить ферзя ни в одной из колонок, решения нет


# Основная функция для запуска решения
n = 8  # Размер доски (8 x 8)
board = [[False] * n for _ in range(n)]  # Создаем доску, заполненную False (пустыми клетками)
solve_n_queens(board, 0, n)  # Запускаем решение с первой строки (row = 0)
```
